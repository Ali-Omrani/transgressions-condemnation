```python
# open a file, where you stored the pickled data
file = open('../../data/5_mil_7days_metoo.p', 'rb')

# dump information to that file
pred_data = pickle.load(file)

# close the file
file.close()
```

```python
pred_data.to_csv("../../data/5_mil_7days_metoo.csv", index=False)
```

```python
user_ids = pred_data.user_id
user_ids
```

```python
import numpy as np
import pandas as pd
unique_user_ids = user_ids.unique()
n = 20
for num, item in enumerate(np.array_split(unique_user_ids, n)):
    df = pd.DataFrame(data=item, columns=["user_id"])
    df.to_csv("../../data/user_ids/user_ids_5mil_7days_part_"+str(num)+".csv", index=False)
    print(len(df))
    print(num)
```
