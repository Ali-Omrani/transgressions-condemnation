```python
def get_tweet_ids(file_path):
    df = pd.read_json(file_path, lines=True)
    print(df.columns)
    print("total_records {}".format(len(df)))
    if "label" in df.columns:
        print("label==1 {}".format(sum(df.label)))
    else:
        print("no label col")
#         print(df.columns)
        print("answer==accept {}".format(sum(df.answer=="accept")))
    

    print("-"*30)
    try:
        tweet_ids = [m["tweet_id"] for m in df.meta]
        df["tweet_id"] = tweet_ids
        return tweet_ids, df
    except:
        return []
```

```python
def get_common_tweet_ids(tweet_id_dict):
    common_dict = {}
    for key1 in tweet_id_dict.keys():
        for key2 in tweet_id_dict.keys():
            if key1 == key2:
                continue
            if key1 +"---and---"+ key2 in common_dict.keys() or key2 +"---and---"+ key1 in common_dict.keys():
                continue
            common = []
            for t_id in tweet_id_dict[key1]:
                for t_id_2 in tweet_id_dict[key2]:
                    if t_id == t_id_2:
                        common.append(t_id)

            common_dict[key1+"---and---"+key2] = common
    return common_dict
get_common_tweet_ids(tweet_id_dict)         
```

```python
def uncommon_label_count(df_dict, common_dict):
    for key in common_dict:
        key1, key2 = key.split("---and---")
        common_tweet_ids = np.unique(common_dict[key])
        print("common_dict", len(common_tweet_ids))
        df1 = df_dict[key1]
        print(len(df1[(df1.tweet_id.isin(common_tweet_ids))].tweet_id))
        df2 = df_dict[key2]
        print(len(df2[(df2.tweet_id.isin(common_tweet_ids))].tweet_id))
        
        for tweet_id in common_tweet_ids:
            if len(df1[(df1.tweet_id.isin(common_tweet_ids))]) > 1:
                print(key1, "has more than one", tweet_id)
                print(df1[(df1.tweet_id.isin(common_tweet_ids))].label)
            if len(df2[(df2.tweet_id.isin(common_tweet_ids))]) > 1:
                print(key2, "has more than one", tweet_id)
uncommon_label_count(df_dict, common_dict)
        
    
```
