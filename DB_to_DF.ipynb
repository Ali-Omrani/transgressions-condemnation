### Filtering Data for Anlaysis

```python
from tqdm import tqdm
```

```python
tqdm.pandas()
```

```python
import pickle

with open("./data/condemnation_RTs.p", "rb") as f:
    df = pickle.load(f)
df
```

```python
df["severity_prediction"].value_counts()
```

Dropping unnecessary columns

```python
columns = ["_id", "postedTime", "retweetCount", "favoritesCount",
           "severity_prediction", "condemnation_prediction", 
          "RT_user_id", "user_id", "time_check", "body_target_mentions",
           "quoted_status_target_mentions", "gnip_url_title_mentions"] 
subset = df[columns]
```

Dropping rows that don't have a single valid time_check

```python
def has_valid_time(row):
#     print(row)
    for key in row["time_check"]:
        if row["time_check"][key]["valid"]:
            return True
    return False
```

```python
subset["time_valid"] = subset.progress_apply(lambda row: has_valid_time(row), axis=1)
```

```python
subset = subset[subset["time_valid"]]
```

```python
subset
```

Gathering all targets

```python
def set_targets(row):
    res = []
    for col in ["body_target_mentions", "quoted_status_target_mentions", "gnip_url_title_mentions"]:
#         print(col)
        if not row[col]:
            continue
        for target in row[col]:
            
            if target not in res:
                res.append(target)
    return res
    
```

```python
subset["targets"] = subset.progress_apply(lambda row: set_targets(row), axis=1) 
```

```python
subset = subset.dropna(subset=["severity_prediction", "favoritesCount", "retweetCount"])
```

```python
subset
```

```python
subset.to_csv("./data/filtered_condemnation_RTs.csv", index=False)
```
